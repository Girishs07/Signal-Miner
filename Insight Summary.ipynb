```python
import pandas as pd

df=pd.read_csv('enhanced_first_dataset.csv')
df
```

üöÄExpansion Signal

```python
sales_roles = df[
    (df['department'].isin(['Sales', 'Product'])) & 
    ((df['Go-To-Market Expansion'] == 1) | (df['General Hiring'] == 1))  
]
```

```python
expansion_signals = sales_roles.groupby('company_name').agg({
    'Job_Title': 'count',
    'Go-To-Market Expansion': 'sum',
    'location': lambda x: x.mode()[0]  
}).sort_values('Job_Title', ascending=False)
```

```python
chargebee_roles = sales_roles[sales_roles['company_name'] == 'Chargebee']
print(chargebee_roles[['Job_Title', 'location']])
```

```python
plum_sales = sales_roles[
    (sales_roles['company_name'] == 'Plum Insurance') & 
    (sales_roles['seniority'].isin(['Lead', 'VP']))
]
print(plum_sales[['Job_Title', 'seniority']])
```

üõ† Product Indicators

```python
product_roles = df[
    (df['department'].isin(['Product', 'Design'])) &  
    ((df['New Product Initiative'] == 1) | (df['Design Overhaul'] == 1))  
]
```

```python
plum_product = product_roles[product_roles['company_name'] == 'Plum Insurance']
plum_counts = plum_product['Job_Title'].value_counts()

print(plum_counts)
```

```python
plum_percentages = plum_product['Job_Title'].value_counts(normalize=True) * 100
print(plum_percentages)
```

```python
razorpay_product = product_roles[product_roles['company_name'] == 'Razorpay']
razorpay_counts = razorpay_product['Job_Title'].value_counts()

print(razorpay_counts)
```

üåç Global Focus

```python
yellow_usa_roles = df[(df['Job_Title'].str.contains('Product Specialist|Account Manager|Account Executive', case=False)) & 
                     (df['location'].str.contains('San Mateo, USA'))]

kala_cyprus_roles = df[(df['Job_Title'].str.contains('B2B Sales|Business Development Manager', case=False)) & 
                      (df['location'].str.contains('Nicosia, Cyprus'))]


print(yellow_usa_roles)
print(kala_cyprus_roles)
```
